## Visualizing data

```python
import numpy as np
import matplotlib.pyplot as plt # for visualizations
import random
```

### Overview of X_train data

```python
fig = plt.figure(figsize=(20, 8))
columns = 8
rows = 5

# plot random 8*5 images from training dataset
for i in range(1, columns*rows +1):
    # get image
    img = X_train[random.randint(0, len(X_train))]
    fig.add_subplot(rows, columns, i)
    # add image to plot
    plt.imshow(img)

plt.show()
```

### Looking at masks in train data

```python
fig = plt.figure(figsize=(20, 8))
columns = 8
rows = 5

# plot random 8*5 masks from the dataset
for i in range(1, columns*rows +1):
    # get mask
    img = y_train[random.randint(0, len(y_train))]
    fig.add_subplot(rows, columns, i)
    # visualize mask
    plt.imshow(img)

plt.show()
```

### Example imput image->mask (for augumented data)

```python
fig = plt.figure(figsize=(8,8))
columns = 2
rows = 1

# X_train input image
fig.add_subplot(rows, columns, 1)
plt.title("Input image")
# show first training example (image)
plt.imshow(X_train[0])

# mask
fig.add_subplot(rows, columns, 2)
plt.title("Mask")
# show corresponding mask
plt.imshow(y_train[0])

plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
columns = 2
rows = 1

# X_train input image
fig.add_subplot(rows, columns, 1)
plt.title("Input image")
# show first training example (image)
plt.imshow(X_train[2000])

# mask
fig.add_subplot(rows, columns, 2)
plt.title("Mask")
# show corresponding mask
plt.imshow(y_train[2000])

plt.show()
```
